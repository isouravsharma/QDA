## Polynomial Regression

Polynomial regression is regression where the model fits a polyniomial curve to your data. i.e. non-linear relationship

### Libraries used numpy matplotlib and scikit-learn

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
# Generate some data
np.random.seed(42)
X = (5*np.random.rand(80,1)) # Random Input Variables
y = X**3 + 2*X+np.random.normal(0, 0.5, size = (80,1)) # Polynomial relationship
```

```python
# Visualize data 
plt.scatter(X,y, color = 'blue', label = 'Data')
plt.title("Sample Data")
plt.xlabel("Feature (X)")
plt.ylabel("Feature (y)")
plt.show()
```

```python
# Transform data into polynomial feature
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
```

```python
# fit the linear refression model to the transformed data
model = LinearRegression()
model.fit(X_poly, y)
```

```python
# plotting the linear regression curve

X_grid = np.arange(min(X), max(X), 0.1) # Smooth curve for plotting
X_grid = X_grid.reshape(len(X_grid),1)
y_grid = model.predict(poly.transform(X_grid))

plt.scatter(X, y, color= 'blue', label = 'Data')
plt.plot(X_grid, y_grid, color = 'red', label = 'Polynomial Regression Line')
plt.title("Polynomial Regression Model")
plt.xlabel("Feature (X)")
plt.ylabel("Target (y)")
plt.legend()
plt.show()
```

```python
# Calculate the MSE

y_pred = model.predict(X_poly)
mse = mean_squared_error(y, y_pred)
mse
```
