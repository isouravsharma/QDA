## DBSCAN

### What is DBSCAN?
- DBSCAN groups point based on their density 

- Algorithm parameters: 
1. Epsilon 
2. MinPts

- Points is based into 3 categories:
- 1. Core Points.
  2. Border Points
  3. Noise Points

```python
# Libraries 
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
```

```python
# Create dataset
X, _ = make_blobs(n_samples = 2000, centers =4, cluster_std=1., random_state=42)
```

```python
# add noise
X = np.vstack([X, np.random.rand(100, 2)* 3 -3])
```

```python
# Standardized 
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Visualize the dataset
plt.scatter(X_scaled[:,0], X_scaled[:,1], s = 10, alpha = .5)
plt.title("Dataset with noise")
plt.xlabel("Feature 1 ")
plt.ylabel("Feature 2")
```

## Implementing the DBSCAN

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps = .2, min_samples=20)
labels = dbscan.fit_predict(X_scaled)
```

```python
# Visualize the cluster result 
plt.scatter(X_scaled[:,0], X_scaled[:,1], c = labels,cmap= 'viridis', s = 10, alpha = .5)
plt.title("Dataset with DBSCAN")
plt.xlabel("Feature 1 ")
plt.ylabel("Feature 2")
```
